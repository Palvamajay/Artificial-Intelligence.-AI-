```python
# import the package
from transformers import pipeline
```

```python
# Load sentiment analysis pipeline with distilBERT
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Sample inputs from IMDB reviews
reviews = [
     "An amazing film â€” I enjoyed every moment!",
     "Totally dull and dragged on forever. Not worth watching.",
    "The plot was so-so, but the performances really stood out.",
]
```

```python
# Predict sentiments
results = classifier(reviews)
for review, result in zip(reviews, results):
    print(f"\nReview: {review}\nPrediction: {result['label']} (score: {result['score']:.2f})")
```

```python
# ==========================
# Q2: Text Generation with Transformers (GPT)
# ==========================
```

```python
# import the package
from transformers import pipeline
```

```python
# Load GPT-2 text generation pipeline
generator = pipeline("text-generation", model="gpt2")
```

```python
# Prompt
prompt = "In a futuristic world, a forgotten secret of the universe is finally revealed..."
```

```python
# Generate story
story = generator(prompt, max_length=80, num_return_sequences=1)[0]['generated_text']
print("\nGenerated Story:\n", story)
```
